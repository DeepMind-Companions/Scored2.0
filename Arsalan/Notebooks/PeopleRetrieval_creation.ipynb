# People Retrieval Creation

### The following file is responsible for getting each player's individual stats and to set an average player as well in case we cannot find that player

We will start by importing the necessary libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('classic')
import json
````

We will get the raw data that I have stored from my previous projects from two csv files.

I have also displayed them here 

````python
bowlingStats = pd.read_csv("Resources/playerData/rawdata/bowling.csv")
battingStats = pd.read_csv("Resources/playerData/rawdata/batting.csv")
bowlingStats
````

````python
battingStats
````

#### We will start by replacing the "-" in each dataframe with np.NaN in order to make sure easy computations

For battingStats, we will also remove the * at the end of each high scores in order to make sure we can convert it to integers.

````python
bowlingStats.replace('-', np.NaN, inplace = True)
battingStats.replace("-", np.NaN, inplace = True)
battingStats['HS'] = battingStats['HS'].str.rstrip("*")
````

Next we will be setting the index of each of these players accordingly and naming the index as well.

````python
bowlingStats.index.names = ['S. no']
battingStats.index.names = ['S. no']
````

#### We are going to add the average for battingStats and bowlingStats now.

````python
battingStats.loc[3492, 'Player'] = 'Average'
bowlingStats.loc[3492, 'Player'] = 'Average'
````

Starting with bowling stats

````python
numeric_cols = ['Mat', 'Inns', 'Overs', 'Mdns', 'Runs', 'Wkts', 'Ave', 'Econ', 'SR', '4', '5']
bowlingStats[numeric_cols] = bowlingStats[numeric_cols].apply(pd.to_numeric, errors='coerce')
column_means = bowlingStats.mean()
bowlingStats.fillna(column_means, inplace = True)
bowlingStats
````

Doing the same for the batting stats

````python
numeric_cols = ['Mat', "Inns", "NO", "Runs", "HS", "Ave", "BF", "SR", "100", "50", "0", "4s", "6s"]
battingStats[numeric_cols] = battingStats[numeric_cols].apply(pd.to_numeric, errors='coerce')
mean_col = battingStats.mean()
battingStats = battingStats.fillna(mean_col)
battingStats
````
