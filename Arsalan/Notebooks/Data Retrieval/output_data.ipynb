# Processing the right output for each match

```python
import pandas as pd
import numpy as np
from pathlib import Path
import random
current_dir = Path.cwd()
parent_dir = current_dir.parent.parent
```

```python
innings = pd.read_pickle(parent_dir/"Resources"/"match_data.pkl")
final_data = pd.read_pickle(parent_dir/"Resources"/"final_data.pkl")
```

```python
final_data
```

```python
innings
```

```python
innings_grouped = innings['score'].groupby(['Match Code', 'Country'])
```

```python
output_scores = innings_grouped.apply(max).to_frame()
output_scores.columns = pd.MultiIndex.from_product([['Final Output'], ['Output Score'], ['Score']])
output_scores
```

```python
final_data.reset_index(["Over", "Ball"])
```

```python
final_data_with_output = final_data.merge(output_scores, left_index = True, right_index = True, how = 'left')

final_data_with_output
```

## Dividing the data

#### In order to divide the data, I will be working with its index and select random matches from it :)}

```python
final_data_with_output.index.levshape
df = final_data_with_output
df.index.levshape
```

This Shows there are a total of 1231 Matches. It will hence be a good idea to divide it into the following Data Set:

| Training Set | Dev Set | Test Set |
| --- | --- | --- |
| 1100 | 100 | 31 |

```python
match_codes = df.index.get_level_values(0).unique().to_list()
random.shuffle(match_codes)
match_codes
```

```python
test_set_codes = match_codes[:31]
dev_set_codes = match_codes[31:131]
training_set_codes = match_codes[131:]

print(len(test_set_codes), len(dev_set_codes), len(training_set_codes))
```

```python
test_list = []
dev_list = []
training_list = []

for code in test_set_codes:
    test_list.append(df.loc[code])

for code in dev_set_codes:
    dev_list.append(df.loc[code])
    
for code in training_set_codes:
    training_list.append(df.loc[code])
```

```python
test_data = pd.concat(test_list, keys = test_set_codes)
dev_data = pd.concat(dev_list, keys = dev_set_codes)
training_data = pd.concat(training_list, keys = training_set_codes)
```
