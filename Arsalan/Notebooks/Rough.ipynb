```python
innings_list = []
team_names = []
for index, each in main_df.iterrows():
    team_names.append(each['team'])
    list_ball_df = []
    list_index = []
    runs_scored = 0
    for over_index, over in enumerate(each['overs']):
        
        # Main data (dropping runs)
        each_ball_series = pd.DataFrame(over['deliveries'])
        each_ball_series = each_ball_series.drop(['runs'], axis = 1)
        
        # Data For the runs scored
        each_ball_series_add = pd.DataFrame([runs['runs'] for runs in over['deliveries']])
        each_ball_series_add.rename(columns = {'batter': 'batter runs', 'extras': 'extra runs', 'total': 'total runs'}, inplace = True)
        each_ball_series_add['score'] = each_ball_series_add['total runs'].cumsum() + runs_scored
        
        
        ball_df = pd.concat([each_ball_series, each_ball_series_add], axis = 1)
        list_ball_df.append(ball_df)
        list_index.append(over_index)
        
        runs_scored += each_ball_series_add['total runs'].sum()
        
    inning_df = pd.concat(list_ball_df, keys = list_index)
    inning_df['out'] = 0
    inning_df['out'].mask(inning_df['wickets'].notna(), 1, inplace = True)
    inning_df['Total Wickets'] = inning_df['out'].cumsum()
    innings_list.append(inning_df)
    # print(inning_df.shape)
    # print(inning_df['wickets'][inning_df['wickets'].notna()])
    # for each in inning_df['wickets'][inning_df['wickets'].notna()]:
       #  print(each)
both_inning_df = pd.concat(innings_list, keys = team_names)
print(both_inning_df)
        
        
```
