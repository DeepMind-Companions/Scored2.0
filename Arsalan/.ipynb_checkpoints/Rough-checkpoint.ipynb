```python
allSeries = []
total_runs = 0
for each_over in main_df.loc[0]['overs']:
    for index, each_delivery in enumerate(each_over['deliveries']):
        total_runs = total_runs + each_delivery['runs']['total']
        newSeries = pd.Series([each_over['over'], 
                                               each_over['over'] * 6 + index + 1,
                                               each_delivery['batter'],
                                                each_delivery['bowler'],
                                                 each_delivery['non_striker'],
                                                  each_delivery['runs']['batter'],
                                                   each_delivery['runs']['extras'],
                                                    each_delivery['runs']['total'],
                                                      total_runs])
        allSeries.append(newSeries)
        print(newSeries)
```

```python
deliveries = pd.DataFrame(allSeries)
```

```python
deliveries.columns = ['over', 'balls', 'batter', 'bowler', 'non_striker', 'runsB', 'runsE', 'runsT', 'Total Runs']
deliveries
```
